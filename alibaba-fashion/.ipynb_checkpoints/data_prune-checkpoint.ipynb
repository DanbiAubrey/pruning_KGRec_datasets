```python
from collections import defaultdict

merged_dict = defaultdict(list)
with open('merged_file.txt', 'r') as file:
    for line in file:
        user, item = map(int, line.split())
        merged_dict[user].append(item)
        
print(len(merged_dict))
```

```python
keys_to_remove = []
for key in merged_dict:
    if len(merged_dict[key]) < 20:
        keys_to_remove.append(key)
```

```python
new_dict = {key: value for key, value in merged_dict.items() if key not in keys_to_remove}
```

```python
print("user number of pruned interaction dataset:{}".format(len(new_dict)))
```

```python
inter = sum([len(item_lst) for item_lst in new_dict.values()])
print("interaction number of pruned interaction dataset:{}".format(inter))
```

```python
import random

test_dict = defaultdict(list)

for key in new_dict.keys():
    random_items = random.sample(new_dict[key], 7)
    
    test_dict[key]+=random_items
    new_dict[key] = [item for item in new_dict[key] if item not in random_items]

trn_inter = sum([len(item_lst) for item_lst in new_dict.values()])
print("interaction number of pruned interaction dataset:{}".format(trn_inter))
tst_inter = sum([len(item_lst) for item_lst in test_dict.values()])
print("interaction number of pruned interaction dataset:{}".format(tst_inter))
print(len(test_dict))
```

```python
trn_lst = []

for key in new_dict.keys():
    for v in new_dict[key]:
        trn_lst.append([key, v])
print(len(trn_lst))
```

```python
tst_lst = []

for key in test_dict.keys():
    for v in test_dict[key]:
        tst_lst.append([key, v])
print(len(tst_lst))
```

```python
with open('pruned_data/test.txt', 'w') as file:
    for row in tst_lst:
        file.write("\t".join(map(str, row)) + '\n')
```
